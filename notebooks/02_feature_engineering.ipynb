```python
import pandas as pd
import numpy as np
```

```python
excel_file = "../data/EduPro Online Platform.xlsx"
```

```python
users = pd.read_excel(excel_file,sheet_name="Users")

courses = pd.read_excel(excel_file,sheet_name="Courses")

transactions = pd.read_excel(excel_file,sheet_name="Transactions")

teachers = pd.read_excel(excel_file,sheet_name="Teachers")
```

```python
master = transactions.merge(
    courses,
    on="CourseID",
    how="left"
)
```

```python
master.head()
```

```python
master = master.merge(
    users,
    on="UserID",
    how="left"
)
```

```python
master.head()
```

```python
master.shape
```

```python
total_courses = master.groupby(
    "UserID"
)["CourseID"].count()
```

```python
total_courses.head()
```

```python
total_spending = master.groupby(
    "UserID"
)["Amount"].sum()
```

```python
total_spending.head()
```

```python
category_diversity = master.groupby(
    "UserID"
)["CourseCategory"].nunique()
```

```python
category_diversity.head()
```

```python
avg_rating = master.groupby(
    "UserID"
)["CourseRating"].mean()
```

```python
avg_rating.head()
```

```python
preferred_category = (
    master.groupby("UserID")
    ["CourseCategory"]
    .agg(lambda x: x.mode()[0])
)
```

```python
preferred_level = (
    master.groupby("UserID")
    ["CourseLevel"]
    .agg(lambda x: x.mode()[0])
)
```

```python
learner_profile = pd.DataFrame({
    "TotalCourses": total_courses,
    "TotalSpending": total_spending,
    "CategoryDiversity": category_diversity,
    "AvgRating": avg_rating,
    "PreferredCategory": preferred_category,
    "PreferredLevel": preferred_level
})
```

```python
learner_profile.head()
```

```python
learner_profile.to_csv(
    "../outputs/learner_profile.csv"
)
```
